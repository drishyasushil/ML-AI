```python
# Separate the dataset into features (X) and the target variable (y)

x = data.drop("price_range", axis=1)
y = data["price_range"]
```

### Standardize (normalize) the feature variables using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x= scaler.fit_transform(x)
```

```python
x
```

```python
# Split the dataset into training and testing sets

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
# Convert target variables to categorical format for multi-class classification

y_train = to_categorical(y_train, num_classes=4)
y_test = to_categorical(y_test, num_classes=4)
```

```python
# Create a Sequential neural network model

model = Sequential()

# Add 2 hidden layers with 128 and 64  neurons and ReLU activation
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
# Add the output layer with 4 neurons (one for each class) and softmax activation
model.add(Dense(4, activation="softmax"))  # Output layer with 4 classes
```

```python
# Compile the neural network model

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
# Train the neural network model

history=model.fit(x_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the neural network model on the testing data
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy}")
```
